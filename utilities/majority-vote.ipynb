```python
import json
import numpy as np
from PIL import Image

def load_annotation(file_path):
    data = []

    with open (file_path, 'r', encoding="utf-8") as file:
        data = json.load(file)    
    return data

def load_image(image_path, size):
    image = Image.open(image_path).convert("RGB").resize(size=size)
    return np.array(image)

annotation_path = "../data/vimmsd-private-test.json"

data = load_annotation(annotation_path)
```

```python
import json

phobert_large_path = "../data/submissions/submission_ver0.7.json"
phobert_large_emojis_path = "../data/submissions/submission_ver0.8.json"
roberta_base_path = "../data/submissions/submission_ver0.9.json"
bert_base_uncased_path = "../data/submissions/submission_ver0.10.json"
phobert_large_emojis_all_path = "../data/submissions/submission_ver0.11.json"

with open(phobert_large_path, 'r') as file:
    phobert_large = json.load(file)
    phobert_large = phobert_large["results"]
print("Length of phobert_large:", len(phobert_large))

with open(phobert_large_emojis_path, 'r') as file:
    phobert_large_emojis = json.load(file)
    phobert_large_emojis = phobert_large_emojis["results"]
print("Length of phobert_large_emojis:", len(phobert_large_emojis))

with open(roberta_base_path, 'r') as file:
    roberta_base = json.load(file)
    roberta_base = roberta_base["results"]
print("Length of roberta_base:", len(roberta_base))

with open(bert_base_uncased_path, 'r') as file:
    bert_base_uncased = json.load(file)
    bert_base_uncased = bert_base_uncased["results"]
print("Length of bert_base_uncased:", len(bert_base_uncased))

with open(phobert_large_emojis_all_path, 'r') as file:
    phobert_large_emojis_all = json.load(file)
    phobert_large_emojis_all = phobert_large_emojis_all["results"]
print("Length of phobert_large_emojis_all:", len(phobert_large_emojis_all))
```

```python
result = {}

for id in phobert_large:
    e_phobert_large = phobert_large[id]
    e_phobert_large_emojis = phobert_large_emojis[id]
    e_roberta_base = roberta_base[id]
    e_bert_base_uncased = bert_base_uncased[id]
    e_phobert_large_emojis_all = phobert_large_emojis_all[id]
    
    entries = [e_phobert_large, e_phobert_large_emojis, e_roberta_base, e_bert_base_uncased, e_phobert_large_emojis_all]
    
    if ("image-sarcasm" in entries):
        result_cur = "image-sarcasm"
    else:
        result_cur = max(set(entries), key=entries.count)
    result[id] = result_cur
```

```python
submission = {"results": {}, "phase": "test"}

for id in result:
    # id = data["key"]
    # label = data["label"]
    submission["results"][id] = result[id]
    
with open("../data/submissions/submission_final.json", "w") as outfile: 
    json.dump(submission, outfile, indent=4)
```
